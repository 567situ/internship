###  1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views

```python
driver=webdriver.Chrome()
```

```python
driver.get(" https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos ")
```

```python
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]
```

```python
rank_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in rank_tags:
    ranks=i.text
    rank.append(ranks)
    
name_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]') 
for i in name_tags:
    names=i.text
    name.append(names)
    
artist_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')  
for i in artist_tags:
    artists=i.text
    artist.append(artists)
    
upload_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]') 
for i in upload_tags:
    uploads=i.text
    upload_date.append(uploads)
    
view_tags=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]') 
for i in view_tags:
    view=i.text
    views.append(view)
```

```python
print(len(rank),len(name),len(artist),len(upload_date),len(views))
```

```python
import pandas as pd
df=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Upload_Date':upload_date,'Views':views})
df
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https:/www.billboard.com/")
```

```python
hot_100=driver.find_element(By.XPATH,'//div[@class="main-sub-menu-container-esp js-hide-when-sticky lrv-u-flex lrv-u-background-color-white u-height-45"]/div/nav/ul/li[4]')
					
hot_100.click()
```

```python
song=[]
artist=[]
last_week_rank=[]
peak_rank=[]
weak_on_board_note=[]
```

```python
import pandas as pd
df=pd.DataFrame({'Song':song,'Artist':artist,'Last_week_rank':last_week_rank,'Peak_rank':peak_rank,'Wak_on_board_note':weak_on_board_note})
df
```

B) Year span 
C) Genre 
D) Run time 
E) Ratings 
F) Votes

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.imdb.com/list/ls095964455/ ")
```

```python
name=[]
year_span=[]

run_time=[]
rating=[]
vote=[]
gener=[]
```

```python
name_tags=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/h3/a')
for i in name_tags:
    names=i.text
    name.append(names)
    
year_tags=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/h3/span[2]') 
for i in year_tags:
    years=i.text
    year_span.append(years)
    
run_tags=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p/span[3]')
for i in run_tags:
    runs=i.text
    run_time.append(runs)
    
rating_tags=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/div/div[1]')
for i in rating_tags:
    ratings=i.text
    rating.append(ratings)
    
vote_tags=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[4]/span[2]') 
for i in vote_tags:
    votes=i.text
    vote.append(votes)
    
gener_tags=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p/span[5]')  
for i in gener_tags:
    geners=i.text
    gener.append(geners)
    
    
```

```python
import pandas as pd
df=pd.DataFrame({'Name':name,'Year_Span':year_span,'Run_Time':run_time,'Rating':rating,'Votes':vote,'Gener':gener})
df
```

### 2. Scrape the details team India’s international fixtures from bcci.tv.

You need to find following details: 
A) Series 
B) Place 
C) Date 
D) Time 
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver=webdriver.Chrome()
```

```python
driver.get(" https://www.bcci.tv/")
```

```python
place=[]
```

```python
place_tags=driver.find_elements(By.XPATH,'//div[@class="match-place ng-scope"]')
for i in place_tags[0:6]:
    plaes=i.text
    place.append(plaes)
    
```

```python
series=[]
```

```python
series_tags=driver.find_elements(By.XPATH,'//div[@class="match-card-middle__inner d-flex justify-content-between"]')
for i in series_tags[0:6]:
    seriess=i.text
    series.append(seriess)
```

```python
time=[]
time_tags=driver.find_elements(By.XPATH,'//div[@class="match-time no-margin ng-binding"]')
for i in time_tags:
    times=i.text
    time.append(times)
    
```

```python
date=[]
date_tags=driver.find_elements(By.XPATH,'//div[@class="match-dates ng-binding"]')
for i in date_tags:
    dates=i.text
    date.append(dates)
```

```python
len(series)
```

```python
import pandas as pd
df=pd.DataFrame({'SERIES':series,'PLACE':place,'DATE':date,'TIME':time})
df
```
